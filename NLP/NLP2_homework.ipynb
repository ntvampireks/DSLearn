```python
import itertools
import string

import numpy as np
import umap
from nltk.tokenize import WordPunctTokenizer

from matplotlib import pyplot as plt

from IPython.display import clear_output
import pandas as pd
```

```python
#data = list(open("./labeled.csv", encoding="utf-8"))
df = pd.read_csv("./labeled.csv")
```

```python
data = list(df["comment"])
```

```python
tokenizer = WordPunctTokenizer()

print(tokenizer.tokenize(data[0]))
```

```python
data_tok = []
for i in range(len(data)): 
    data_tok.append(tokenizer.tokenize(data[i].lower()))
assert all(isinstance(row, (list, tuple)) for row in data_tok), "please convert each line into a list of tokens (strings)"
assert all(all(isinstance(tok, str) for tok in row) for row in data_tok), "please convert each line into a list of tokens (strings)"
is_latin = lambda tok: all('a' <= x.lower() <= 'z' for x in tok)
assert all(map(lambda l: not is_latin(l) or l.islower(), map(' '.join, data_tok))), "please make sure to lowercase the data"

```

```python
dict_words = {}
stopwords=[",",".","и", "в", "не", "-", "на", "что", "а"]


for token in data_tok:
    for word in token: 
        if word not in stopwords:
            if word in dict_words.keys(): 
                dict_words[word] += 1
            else:
                dict_words[word] = 1

import pandas as pd
df = pd.DataFrame.from_dict(dict_words, orient='index')
df
df.sort_values(0, ascending = False).head(10)
```

```python
import heapq

dict_words = {}

for token in data_tok:
    for word in token: 
        if word not in stopwords:
            if word in dict_words.keys(): 
                dict_words[word] += 1
            else:
                dict_words[word] = 1

most_freq = heapq.nlargest(600, dict_words, key=dict_words.get)

sentence_vectors = []
for sentence in data_tok:
    sentence_tokens = sentence
    sent_vec = []
    for token in most_freq:
        if token in sentence_tokens:
            sent_vec.append(1)
        else:
            sent_vec.append(0)
    sentence_vectors.append(sent_vec)
```

```python
df1 = pd.DataFrame(sentence_vectors, columns = most_freq) 
df1
```
