```python
#df = pd.json_normalize(objects_list)
df = pd.DataFrame()
df_metro = pd.DataFrame()
df_photos = pd.DataFrame()
#Пробегаемся по списку с айдишниками объектов, запрашиваем информацию, копируем в датафрейм
for i in objids:
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{i}'
    json = requests.get(url).json()
    main = pd.json_normalize(json['data'])
    df = pd.concat([df, main]) 
    metro = pd.json_normalize(json['data'], 
                  record_path = ['objectTransportInfo'],
                  meta = [['id'],'metroStationNm', 'isMetroStationDistWalking', 
                          'metroStationDistTime', 'metroStationDistLength',
                         'metroLineColorHexCode', 'metroLineColorsHexCode' ],
                  errors = 'ignore',
                  record_prefix = 'objectTransportInfo.'
                 ).dropna(axis=1, how='all')
    df_metro = pd.concat([df_metro, metro])
    photos = pd.json_normalize(json['data'], 
                    record_path = ['photoRenderDTO'])
    df_photos = pd.concat([df_photos, photos])
#SQLLitte по некоторым причинам не может записать поля которые начинаются и заканчиваются на []
#пока удаляем
del df['objectTransportInfo']
del df['photoRenderDTO']
del df['metro.colors']
del df_metro['objectTransportInfo.metroLineColorsHexCode']
df.to_excel("output.xlsx")
df.to_pickle("test.pkl")
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
df.to_sql('Main', engine)
df_metro.to_sql('Metro_near_objects', engine)
df_photos.to_sql('Objects_photos', engine)
```
